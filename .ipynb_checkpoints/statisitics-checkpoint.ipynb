```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# === CONFIG ===
logfile = "/home/phanim/harshitrawat/summer/results/mace_T1_finetune_h200_cn10_run-21_train.txt"  # 🔁 uptil epoch 15

# === PARSE LOG FILE ===
records = []
with open(logfile, "r") as f:
    for line in f:
        try:
            record = json.loads(line)
            if record.get("mode") == "eval" and "epoch" in record:
                records.append(record)
        except json.JSONDecodeError:
            continue

df = pd.DataFrame(records)

# === EXPLANATION OF METRICS ===
explanations = {
    "loss": "Total loss used in optimization (combined energy + force)",
    "mae_e": "Mean Absolute Error of total energy (eV)",
    "mae_e_per_atom": "MAE of energy per atom (eV/atom)",
    "rmse_e": "Root Mean Squared Error of total energy (eV)",
    "rmse_e_per_atom": "RMSE of energy per atom (eV/atom)",
    "mae_f": "Mean Absolute Error of atomic forces (eV/Å)",
    "rmse_f": "Root Mean Squared Error of forces (eV/Å)",
    "p95_e_per_atom": "95th percentile of energy error per atom (eV)",
    "p95_f": "95th percentile of force error (eV/Å)",
    "epoch": "Training epoch number",
    "mode": "eval or train (we only analyze eval here)"
}

# === PLOT METRICS ===
metrics = [col for col in df.columns if col not in ["epoch", "mode"]]

for metric in metrics:
    plt.figure()
    plt.plot(df["epoch"], df[metric], marker='o')
    plt.title(f"{metric} over Epochs\n{explanations.get(metric, '')}")
    plt.xlabel("Epoch")
    plt.ylabel(metric)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
%matplotlib inline
import json
import pandas as pd
import matplotlib.pyplot as plt

# ===== CONFIG =====
log_file = "/home/phanim/harshitrawat/summer/results/mace_T1_finetune_h200_cn10_run-21_train.txt"  # 🔁 Replace with your path

# ===== PARSE AND CONVERT UNITS =====
eval_data = []
with open(log_file, "r") as f:
    for line in f:
        try:
            entry = json.loads(line)
            if entry.get("mode") == "eval" and entry.get("epoch") is not None:
                eval_data.append({
                    "epoch": int(entry["epoch"]),
                    "rmse_e (meV)": 1000 * entry["rmse_e"],
                    "rmse_e_per_atom (meV)": 1000 * entry["rmse_e_per_atom"],
                    "rmse_f (meV/Å)": 1000 * entry["rmse_f"]
                })
        except json.JSONDecodeError:
            continue

df = pd.DataFrame(eval_data).sort_values("epoch")

# ===== SUMMARY STATS =====
stats = df[["rmse_e (meV)", "rmse_e_per_atom (meV)", "rmse_f (meV/Å)"]].agg(["min", "max", "std"])
print("📊 RMSE Summary Stats (units converted):\n")
print(stats)

# If you want to use df later:
# df.to_csv("parsed_eval_metrics.csv", index=False)
plot_configs = [
    ("rmse_e (meV)", "RMSE Energy (meV)", "rmse_energy_plot_T1training.png"),
    ("rmse_e_per_atom (meV)", "RMSE Energy per Atom (meV)", "rmse_energy_per_atom_plot_T1training.png"),
    ("rmse_f (meV/Å)", "RMSE Forces (meV/Å)", "rmse_forces_plot_T1training.png"),
]

for column, title, filename in plot_configs:
    plt.figure(figsize=(8, 5))
    plt.plot(df["epoch"], df[column], marker="o")
    plt.xlabel("Epoch")
    plt.ylabel(column)
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.close()
    plt.show()
```
